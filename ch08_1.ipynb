# 합성곱 신경망의 구성 요소

**합성곱(Convolutional neural network, CNN)**

neuron = filter(뉴런의 개수를 이야기할 때) = kernel(입력에 곱하는 가중치)

특성 맵(feature map): 합성곱 계산을 통해 얻은 출력.

**케라스 합성곱 층**

합성곱 신경망 : 1개 이상의 합성곱 층을 쓴 인공신경망

**패딩과 스트라이드**

패딩(padding): 입력 배열의 주위를 가상의 원소로 채우는 것.

세임 패딩(same padding): 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩 하는 것.

밸리드 패딩(valid padding): 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우

스트라이드(stride): 합성곱 연산을 할 때의 이동의 크기, 스트라이드가 커지면 만들어지는 특성 맵의 크기는 더 작아진다.

풀링(pooling): 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할.
- max pooling
- average pooling
